```python
import numpy as np
import pandas as pd
from ast import literal_eval
```

```python
df = pd.read_csv('movie_all.csv',encoding ='utf-8')
```

```python
df.head()
```

```python
df.info()
```

```python
# drop unnecessary columns
df = df.drop(columns=['Unnamed: 0', 'actor1_gender', 'actor2_gender', 'actor3_gender', 'actor4_gender', 'actor5_gender', 'actor_number', 'director_gender', 'director_number', 'producer_number'])
```

```python
# replace string 'none' in some cells with np.NaN(nontype)
df = df.replace('none', np.NaN)
```

```python
df.info()
```

```python
# make 'release_date' column be datetime type
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
df.head()['release_date']
```

```python
# add release_year column
df['release_year'] = df['release_date'].dt.year
df.head(1)
```

```python
df.columns.get_loc("actor1_name")
```

```python
# make actor1_name to actor5_name in one
# and drop nan 

alist = []
for ind, value in df.iloc[:, 22:27].iterrows():
    
    # set these five actors 
    
    ok = set(value.squeeze()) # for the first row is {nan, 'Eetu Hilkamo', 'Matti Pellonpää', 'Susanna Haavisto', 'Turo Pajala'}
    
    # delect nan in the set
    
    ok = {x for x in ok if x==x} 
    # for the first row is 
    # {'Susanna Haavisto', 'Turo Pajala', 'Matti Pellonpää', 'Eetu Hilkamo'}
    
    alist.append(ok)
alist
```

```python
# add alist into one column in df

df_2=df.assign(actor_set=alist)
df_2
```

```python
df_2['actor_num']  = df_2['actor_set'].str.len()
```

```python
df_2.head()
```

```python
# for further group by, order the actor set
# ex: {A,B} is the same as {B,A} 
actor_list = []
for ind, value in df_2.iterrows():
    ok = sorted(value['actor_set'])
    actor_list.append(ok)
actor_list
```

```python
# stored actor_list as a tuple in a column
df_2=df_2.assign(actor_list=actor_list)
df_2['actor_list'] = df_2['actor_list'].apply(tuple)
df_2.head()
```

```python
df_2.to_csv('movie_final.csv')
```

# 分組輸出CSV檔

# 1.Node = Actor, Edge = Director

### [參考](https://stackoverflow.com/questions/54944344/groupby-and-append-lists-and-strings)

```python
# groupby 導演及年份，因actor_list欄位一個cell可能含有多個elements, 
# 所以groupby時actor_list是用sum的方式將這些elements合併
# ex: ('Jiaxuan Li', 'Tang Wei, Tian Yu') + ('Tao Zeru',) = ('Jiaxuan Li', 'Tang Wei, Tian Yu', 'Tao Zeru')
d1 = df_3.replace('',None).groupby(['director_name', 'release_year']).agg({'actor_list':'sum'}).reset_index()
d1
```

```python
d1.to_csv('1_N_Acotor_E_Director.csv')
```

# 2. Node = Movie, Edge = Actor

```python
df_3.head(2)
```

```python
df_3.iloc[:, 22:27].head()
```

### [參考](https://stackoverflow.com/questions/46622869/pandas-groupby-column-a-and-make-lists-of-tuples-from-other-columns)

```python
# 處理actor 1並groupby actor 1
# 因title不像第一組actor為多個element在一個cell，故這裡lambda x: tuple(x.values)對單個elemen一個個做合併並設為tuple
dd1 = df_3.groupby(['actor1_name', 'release_year'])['title'].apply(lambda x: tuple(x.values)).reset_index()
dd1= dd1.rename(columns = {'actor1_name':'actor_name'})
dd1.head()
```

```python
# ex:可以看到合併了50 Cent這演員在2005年的兩部電影==(Get Rich or Die Tryin', Beef III)
dd1[dd1['actor_name'] == '50 Cent']
```

```python
#一次處理for actor 1 to actor 5並垂直合併這五個dataframe
actors = ['actor1_name', 'actor2_name', 'actor3_name', 'actor4_name', 'actor5_name']
d2 = pd.DataFrame()
for actor in actors:
    
    pre_d2 = df_3.groupby([actor, 'release_year'])['title'].apply(lambda x: tuple(x.values)).reset_index()
    pre_d2 = pre_d2.rename(columns = {actor:'actor_name'})
    print(pre_d2.shape[0])
    d2 = d2.append(pre_d2)
```

```python
sum([46701,
88607,
81131,
74152,
65391])
```

```python
d2.shape
```

```python
d2 = d2.reset_index(drop = True)
d2
```

```python
# 因垂直合併可能會有重複的actor，故再groupby一次
# 這次因title為多個element，故用像第一組groupby用 sum'的方式合併這些elements
# ex: ('Jiaxuan Li', 'Tang Wei, Tian Yu') + ('Tao Zeru',) = ('Jiaxuan Li', 'Tang Wei, Tian Yu', 'Tao Zeru')
dd2 = d2.replace('',None).groupby(['actor_name', 'release_year']).agg({'title':'sum'}).reset_index()
dd2
```

```python
dd2.to_csv('2_N_Movie_E_Actor.csv')
```
